```python
import numpy as np, string, utils, preprocess
import nltk
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
book = utils.load_text("books/king-james-bible-processed.txt")
```

```python
print(book[:100])
```

```python
book_lems = preprocess.words_lems(book, lower=True)
```

```python
fdist = nltk.FreqDist(book_lems)
fdist.most_common(10)
fdist.plot(50,cumulative=True)
```

```python
text2 = nltk.Text(nltk.corpus.gutenberg.words('austen-sense.txt'))
fdist2 = nltk.FreqDist(text2)
```

```python
#fdist2.freq('charles')
l = []
for w in fdist:
    #print(w)
    #print(fdist2.freq(w))
    l.append((w, fdist.freq(w) / (fdist2.freq(w) + 1e-10)))
```

```python
#nltk.FreqDist(l).plot(50,cumulative=True)
l = sorted(l, key=lambda t: int(t[1]), reverse=True)

arr = np.array(l)[0:10]
arr = [list(d[:10]) for d in zip(*l)]
sns.barplot(x=arr[0], y=arr[1])
plt.plot()
plt.show()
```
